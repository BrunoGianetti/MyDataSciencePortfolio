```python
import matplotlib.pyplot as plt

# Transformar a coluna "data" para datetime
resumo_mundial["data"] = pd.to_datetime(resumo_mundial["data"])

# Criar um gráfico linear
plt.figure(figsize=(15, 6))
plt.plot(resumo_mundial["data"], resumo_mundial["consumo_total_de_energia_primaria"], label="Consumo de energia elétrica mundial")
plt.plot(resumo_mundial["data"], resumo_mundial["producao_total_de_energia_primaria"], label="Produção de energia elétrica mundial")

# Adicionar uma legenda
plt.legend()

# Limpar os dados do eixo X para leitura
plt.xticks(rotation=45)

# Adicionar um título e labels para eixo X e Y
plt.title("Produção e consumo de energia elétrica mundial")
plt.xlabel("Data")
plt.ylabel("Quadrilhões de BTU")

# Exibir o gráfico
plt.show()
```

```python
# Agrupar os dados por continente e calcular a soma da coluna "producao_total"
producao_total = producao.groupby("continente")["producao_total"].sum()

# Ordenar os dados do menor para o maior
producao_total = producao_total.sort_values()

# Criar um novo dataframe com os dados agrupados e ordenados
producao_total = pd.DataFrame(producao_total)

# Criar um gráfico de barras horizontal
plt.figure(figsize=(7, 7))
plt.barh(producao_total.index, producao_total["producao_total"], color="#6A58CC")

# Adicionar um título e labels para o eixo X
plt.title("Produção total de energia por continente")
plt.xlabel("Quadrilhões de BTU")

# Exibir o gráfico
plt.show()
```

```python
# Agrupar os dados por continente e calcular a soma da coluna "producao_total"
producao_total = producao.groupby("continente")["producao_total"].mean()

# Ordenar os dados do menor para o maior
producao_total = producao_total.sort_values()

# Criar um novo dataframe com os dados agrupados e ordenados
producao_total = pd.DataFrame(producao_total)

# Criar um gráfico de barras horizontal
plt.figure(figsize=(7, 7))
plt.barh(producao_total.index, producao_total["producao_total"], color="#6A58CC")

# Adicionar um título e labels para o eixo X
plt.title("Média da podução total de energia por continente")
plt.xlabel("Quadrilhões de BTU")

# Exibir o gráfico
plt.show()
```

```python
# Criar um novo dataframe com apenas os anos de "2007","2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"
producao_anos = producao[["pais", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]]

# Somar o total de energia produzida em cada país
producao_anos["producao_total"] = producao_anos[["2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]].sum(axis=1)

# Ordenar os dados pelo valor da coluna "producao_total" em ordem decrescente
producao_anos = producao_anos.sort_values("producao_total", ascending=False)

# Selecionar as primeiras 10 linhas do dataframe
producao_anos = producao_anos.head(10).round(2)

# Exibir o dataframe
producao_anos[['pais','producao_total']]
```

```python
import plotly.graph_objects as go

# 1. Criar um novo DataFrame com as colunas de interesse
colunas_anos = ["2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]
producao_pais = producao_anos[colunas_anos + ["pais"]]

# 2. Criar a plotagem usando plotly.graph_objects
fig = go.Figure()

for pais in producao_pais["pais"]:
    dados_pais = producao_pais[producao_pais["pais"] == pais]
    fig.add_trace(go.Scatter(
        x=colunas_anos,
        y=dados_pais[colunas_anos].values.tolist()[0],
        mode='lines+markers',
        name=pais
    ))

fig.update_layout(
    title='Histórico de Produção de Energia por País (2007-2021)',
    xaxis_title='Ano',
    yaxis_title='Produção de Energia (quadrilhões de BTU)',
)

fig.show()
```

```python
import statsmodels.api as sm

# 2. Criar um novo DataFrame com as colunas de interesse
producao_tipos = resumo_mundial[["data", "producao_total_de_combustiveis_fosseis", "producao_de_energia_nuclear", "producao_total_de_energia_renovavel"]]

# 3. Utilizar seasonal_decompose para decompor as séries temporais
producao_tipos.set_index("data", inplace=True)
decomposed_combustiveis = sm.tsa.seasonal_decompose(producao_tipos["producao_total_de_combustiveis_fosseis"], model='additive')
decomposed_nuclear = sm.tsa.seasonal_decompose(producao_tipos["producao_de_energia_nuclear"], model='additive')
decomposed_renovavel = sm.tsa.seasonal_decompose(producao_tipos["producao_total_de_energia_renovavel"], model='additive')

# 4. Criar a plotagem usando plotly
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=producao_tipos.index,
    y=decomposed_combustiveis.trend,
    mode='lines',
    name='Produção de Combustíveis Fósseis (Tendência)'
))


fig.add_trace(go.Scatter(
    x=producao_tipos.index,
    y=decomposed_nuclear.trend,
    mode='lines',
    name='Produção de Energia Nuclear (Tendência)'
))


fig.add_trace(go.Scatter(
    x=producao_tipos.index,
    y=decomposed_renovavel.trend,
    mode='lines',
    name='Produção de Energia Renovável (Tendência)'
))

fig.update_layout(
    title='Tendência de Produção de Energia por Tipo',
    xaxis_title='Data',
    yaxis_title='Produção de Energia',
    legend_title='Tipo de Produção',
    plot_bgcolor='white'  # Definir o fundo do gráfico como branco
)

fig.show()
```
